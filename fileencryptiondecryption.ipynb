```python
from cryptography.fernet import Fernet
import os

def generate_key():
    key = Fernet.generate_key()
    with open("secret.key", "wb") as key_file:
        key_file.write(key)

def load_key():
    return open("secret.key", "rb").read()

def encrypt_file(file_name):
    # Load the key
    key = load_key()
    f = Fernet(key)
    
    with open(file_name, "rb") as file:
        file_data = file.read()

    encrypted_data = f.encrypt(file_data)

    with open(f"encrypted_{file_name}", "wb") as file:
        file.write(encrypted_data)

    print(f"File {file_name} encrypted successfully and saved as encrypted_{file_name}")

def decrypt_file(file_name):
    
    key = load_key()
    f = Fernet(key)

    with open(file_name, "rb") as file:
        encrypted_data = file.read()

    decrypted_data = f.decrypt(encrypted_data)

    with open(f"decrypted_{file_name}", "wb") as file:
        file.write(decrypted_data)

    print(f"File {file_name} decrypted successfully and saved as decrypted_{file_name}")

def main():
    while True:
        print("\nSelect an option:")
        print("1. Generate a new encryption key")
        print("2. Encrypt a file")
        print("3. Decrypt a file")
        print("4. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            generate_key()
            print("Encryption key generated and saved as secret.key")

        elif choice == '2':
            file_name = input("Enter the file name to encrypt: ")
            if os.path.exists(file_name):
                encrypt_file(file_name)
            else:
                print("File not found!")

        elif choice == '3':
            file_name = input("Enter the file name to decrypt: ")
            if os.path.exists(file_name):
                decrypt_file(file_name)
            else:
                print("File not found!")

        elif choice == '4':
            print("Exiting the program.")
            break

        else:
            print("Invalid choice, please try again.")

if __name__ == "__main__":
    main()
```
